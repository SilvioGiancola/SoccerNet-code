```python
import numpy as np

def NonMaxSuppression(Input, window=60):

    res = np.empty(np.size(Input), dtype=bool)
#     print(np.size(Input))
#     print(res.shape)
    for i, value  in enumerate(Input):
        if (i==0 or i ==np.size(Input)-1):
            res[i] = False
            continue
#         print(i,value)
        if value > Input[i-1] and value > Input[i+1]:
#             print (Input[i-1], value, Input[i+1])
            res[i] = True
            for past in range(int(window/2)):
                if (i-past < 0) : continue
                if (res[i-past] == True and Input[i-past] > Input[i]) : 
                    res[i] = False
                    break
                if (res[i-past] == True and Input[i-past] < Input[i]) :
                    res[i-past] = False
                    
    MaxValues = Input[res==True]
    indexes = np.arange(np.size(Input))[res==True]
    return MaxValues, indexes
        
print((Input))
print(NonMaxSuppression(Input,window=2)[0])
print(NonMaxSuppression(Input,window=2)[1])
```

```python
from itertools import islice

def window(seq, n=2):
    "Returns a sliding window (of width n) over data from the iterable"
    "   s -> (s0,s1,...s[n-1]), (s1,s2,...,sn), ...                   "
    it = iter(seq)
    result = tuple(islice(it, n))
    if len(result) == n:
        yield result
    for elem in it:
        result = result[1:] + (elem,)
        yield result
```
